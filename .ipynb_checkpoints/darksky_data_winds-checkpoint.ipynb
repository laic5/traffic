```python
from darksky import forecast
from datetime import datetime as dt
from datetime import date, timedelta
import pandas as pd
```

Using Time Machine Call, we can get the observed weather conditions for a date in the past or future. 

```python
key = "4ff1d8e73db78778122dfacfabf261f2"
Livemore = key, 37.681, -121.775
```

```python
t = dt(2018, 1, 17, 12).isoformat()
live = forecast(*Livemore, t)
mydata = pd.DataFrame(live['hourly']['data'])
mydata['LAT'] = 37.681
mydata['LON'] = -121.775
mydata['YEAR'] = 2018
mydata['MONTH'] = 1
mydata['DAY'] = 17
mydata['HOUR'] = [i for i in range(24)]
```

```python
for i in range(18, 32):  
    t = dt(2018, 1, i, 12).isoformat()
    live = forecast(*Livemore, t)
    tempdata = pd.DataFrame(live['hourly']['data'])
    tempdata['LAT'] = 37.681
    tempdata['LON'] = -121.775
    tempdata['YEAR'] = 2018
    tempdata['MONTH'] = 1
    tempdata['DAY'] = i
    tempdata['HOUR'] = [i for i in range(24)]
    mydata = mydata.append(tempdata)
    
Jan_winds = mydata
```

Save it as csv

```python
Jan_winds.to_csv("Jan_winds.csv", sep=',')
```
