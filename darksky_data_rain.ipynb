```python
from darksky import forecast
from datetime import datetime as dt
from datetime import date, timedelta
import pandas as pd
```

Using Time Machine Call, we can get the observed weather conditions for a date in the past or future. 

```python
key = "4ff1d8e73db78778122dfacfabf261f2"
Fremont = key, 37.542, -121.991
```

```python
t = dt(2018, 1, 1, 12).isoformat()
fre = forecast(*Fremont, t)
mydata = pd.DataFrame(fre['hourly']['data'])
mydata['LAT'] = 37.542
mydata['LON'] = -121.991
mydata['YEAR'] = 2018
mydata['MONTH'] = 1
mydata['DAY'] = 1
mydata['HOUR'] = [i for i in range(24)]
```

```python
for i in range(2, 14):  
    t = dt(2018, 1, i, 12).isoformat()
    fre = forecast(*Fremont, t)
    tempdata = pd.DataFrame(fre['hourly']['data'])
    tempdata['LAT'] = 37.542
    tempdata['LON'] = -121.991
    tempdata['YEAR'] = 2018
    tempdata['MONTH'] = 1
    tempdata['DAY'] = i
    tempdata['HOUR'] = [i for i in range(24)]
    mydata = mydata.append(tempdata)
    
Jan_week = mydata
```

Save it as csv

```python
Jan_week.to_csv("Jan_week.csv", sep=',')
```
